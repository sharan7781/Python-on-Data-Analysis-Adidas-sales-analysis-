Exploratory Data Analysis(EDA)

```python
#exploratory Data Analysis(EDA)
ax = sns.countplot(x= 'Region', data= df)
```

```python
ax = sns.countplot(x= 'Region', data= df)  ## plotting a bar chart for Region and it's count


for bars in ax.containers:
  ax.bar_label(bars)
```

```python
df.groupby(['Region'], as_index=False)['Units Sold'].sum().sort_values(by='Units Sold', ascending=False)
```

```python
sales_Product = df.groupby(['Region'], as_index=False)['Units Sold'].sum().sort_values(by='Units Sold', ascending=False)

sns.barplot(x = 'Region',y = 'Units Sold' ,data = sales_Product)
```

```python
df.columns
```

product vs region

```python
ax = sns.countplot(data = df, x = 'Region', hue = 'Product')
sns.set(rc={'figure.figsize':(17,7)})
for bars in ax.containers:
    ax.bar_label(bars)
```

Retailer vs product

```python
ax = sns.countplot(data = df, x = 'Retailer', hue = 'Product')  ## Retailer vs product
sns.set(rc={'figure.figsize':(30,20)})
for bars in ax.containers:
    ax.bar_label(bars)
```

State vs Product

```python
ax = sns.countplot(data=df, x='State', hue='Product', order=df['State'].value_counts().iloc[:5].index)  ## products in state wise
sns.set(rc={'figure.figsize':(15, 10)})
for bars in ax.containers:
    ax.bar_label(bars)
```
